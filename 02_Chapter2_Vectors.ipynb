## Summary
Chapter ini membahas konsep vektor sebagai dasar dalam linear algebra yang digunakan dalam data science.

## Tujuan

- Melakukan operasi dasar vektor
- Melihat visualisasi sederhana

## Konsep Dasar

Vektor adalah kumpulan angka yang merepresentasikan data.
Dalam Python, vektor biasanya direpresentasikan menggunakan NumPy array.

Contoh:
[1, 2, 3] adalah sebuah vektor 3 dimensi.

```python
import numpy as np

v = np.array([1, 2, 3])
print("Vector:", v)
```

## Operasi Vektor

Beberapa operasi dasar pada vektor:
- Penjumlahan
- Perkalian (dot product)

```python
v1 = np.array([1,2,3])
v2 = np.array([4,5,6])

print("Penjumlahan:", v1 + v2)
print("Dot product:", np.dot(v1, v2))
```

## Visualisasi Vektor

Visualisasi membantu memahami arah dan nilai vektor.

```python
import matplotlib.pyplot as plt

# vektor
v = np.array([3, 4])

# plot vektor sebagai panah
plt.quiver(0, 0, v[0], v[1], angles='xy', scale_units='xy', scale=1)

plt.xlim(0,5)
plt.ylim(0,5)

plt.title("Visualisasi Vektor")
plt.grid()

plt.show()
```

## Kesimpulan

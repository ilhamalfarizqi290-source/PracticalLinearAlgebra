# Chapter 5 - Matrix Operations & Determinant

## Summary
Chapter ini membahas operasi lanjutan pada matriks seperti determinan dan invers yang penting dalam linear algebra dan machine learning.

## Tujuan
- Memahami determinan matriks
- Menghitung invers matriks
- Melihat efek transformasi matriks

## Konsep Dasar

Beberapa konsep penting:
- Determinan: menunjukkan apakah matriks dapat dibalik
- Invers: kebalikan dari matriks
- Matriks digunakan untuk transformasi data

```python
import numpy as np

A = np.array([[1,2],
              [3,4]])

det = np.linalg.det(A)
print("Determinan:", det)
```

```python
inv_A = np.linalg.inv(A)
print("Inverse Matrix:\n", inv_A)
```

```python
import matplotlib.pyplot as plt

# bentuk awal (persegi)
square = np.array([[0,0],[1,0],[1,1],[0,1],[0,0]])

# transformasi matriks
A = np.array([[1,2],
              [3,4]])

transformed_square = square @ A.T

plt.plot(square[:,0], square[:,1], label="Original")
plt.plot(transformed_square[:,0], transformed_square[:,1], label="Transformed")

plt.title("Area Transformation by Matrix")
plt.legend()
plt.grid()

plt.show()
```

```python
identity = np.dot(A, inv_A)
print("A x A^-1:\n", identity)
```

```python
v = np.array([1,1])

v_trans = A @ v
v_back = inv_A @ v_trans

plt.quiver(0,0,v[0],v[1], angles='xy', scale_units='xy', scale=1, label='Original')
plt.quiver(0,0,v_trans[0],v_trans[1], angles='xy', scale_units='xy', scale=1, label='Transformed')
plt.quiver(0,0,v_back[0],v_back[1], angles='xy', scale_units='xy', scale=1, label='Recovered')

plt.xlim(-2,5)
plt.ylim(-2,5)

plt.title("Matrix and Inverse Transformation")
plt.grid()
plt.legend()

plt.show()
```

## Kesimpulan

Determinan menunjukkan perubahan skala akibat matriks.
Invers matriks digunakan untuk mengembalikan transformasi ke bentuk semula.

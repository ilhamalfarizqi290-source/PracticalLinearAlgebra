# Chapter 6 - Matrix Decomposition (QR)

## Summary
QR decomposition memecah matriks menjadi Q (orthogonal) dan R (upper triangular).

## Konsep Utama
- Q orthogonal (QᵀQ = I)
- R segitiga atas
- Digunakan dalam solving system & least squares

```python
import numpy as np

A = np.array([[1,2],
              [3,4]])

Q, R = np.linalg.qr(A)

print("Q:\n", Q)
print("R:\n", R)
```

```python
import matplotlib.pyplot as plt

plt.imshow(Q)
plt.colorbar()
plt.title("Matrix Q Heatmap")
plt.show()
```

QR decomposition memudahkan analisis matriks dan banyak dipakai dalam algoritma numerik.

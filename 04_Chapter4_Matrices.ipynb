Digunakan dalam:

```python
import numpy as np

A = np.array([[1,2,3],
              [4,5,6]])

print(A)
```

```python
# Transpose
A_T = A.T
print("Transpose:\n", A_T)

# Matrix multiplication
B = np.array([[1,0],
              [0,1],
              [1,1]])

result = np.dot(A, B)
print("Hasil perkalian:\n", result)
```

```python
import matplotlib.pyplot as plt

matrix = np.array([[1,2,3],
                   [4,5,6]])

plt.imshow(matrix)
plt.colorbar()

plt.title("Matrix Heatmap Visualization")

for i in range(matrix.shape[0]):
    for j in range(matrix.shape[1]):
        plt.text(j, i, matrix[i, j], ha='center', va='center', color='white')

plt.show()
```

```python
scale_matrix = np.array([[2,0],
                         [0,2]])

vector = np.array([1,1])

transformed = np.dot(scale_matrix, vector)

print("Hasil transformasi:", transformed)
```

```python
v = np.array([1,1])
v_trans = transformed

plt.quiver(0,0,v[0],v[1], angles='xy', scale_units='xy', scale=1, label='Original')
plt.quiver(0,0,v_trans[0],v_trans[1], angles='xy', scale_units='xy', scale=1, label='Transformed')

plt.xlim(0,3)
plt.ylim(0,3)

plt.title("Vector Transformation using Matrix")
plt.grid()
plt.legend()

plt.show()
```

## Kesimpulan

Matriks digunakan untuk merepresentasikan data dan melakukan transformasi dalam machine learning.
Visualisasi membantu memahami bagaimana matriks mengubah vektor.

# Chapter 7 - Matrix Rank & Linear Independence

## Summary
Rank menunjukkan jumlah informasi independen dalam matriks.

## Konsep Utama
- Rank = jumlah baris/kolom independen
- Linear independence

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
A = np.array([[1,2,3],
              [2,4,6],
              [1,1,1]])

rank = np.linalg.matrix_rank(A)
print("Rank:", rank)
```

```python
import matplotlib.pyplot as plt

plt.imshow(A)
plt.colorbar()
plt.title("Matrix Visualization")
plt.show()
```

```python
plt.bar(["Rank","Dimensi"], [rank, A.shape[0]])
plt.title("Rank vs Dimension")
plt.show()
```

```python
plt.bar(["Rank","Dimensi"], [rank, A.shape[0]])
plt.title("Rank vs Dimension")
plt.show()
```

Rank membantu menentukan apakah data redundan atau tidak.

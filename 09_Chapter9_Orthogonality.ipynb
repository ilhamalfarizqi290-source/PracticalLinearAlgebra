# Chapter 9 - Orthogonality

## Summary
Dua vektor orthogonal jika dot product = 0.

## Konsep Utama
- Orthogonal vectors
- Sudut 90°

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
v1 = np.array([1,0])
v2 = np.array([0,1])

dot = np.dot(v1, v2)
print("Dot product:", dot)
```

```python
import matplotlib.pyplot as plt

plt.quiver(0,0,1,0, angles='xy', scale_units='xy', scale=1, label='v1')
plt.quiver(0,0,0,1, angles='xy', scale_units='xy', scale=1, label='v2')

plt.legend(); plt.grid(); plt.title("Orthogonal Vectors"); plt.show()
```

```python
theta = np.linspace(0, np.pi/2, 50)
plt.plot(np.cos(theta), np.sin(theta))
plt.title("90 Degree Relationship")
plt.grid()
plt.show()
```

Orthogonality penting dalam optimasi dan dekomposisi matriks.

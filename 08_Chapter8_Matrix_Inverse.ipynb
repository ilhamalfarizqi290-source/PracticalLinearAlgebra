# Chapter 8 - Matrix Inverse & Identity

## Summary
Inverse mengembalikan matriks ke bentuk identitas.

## Konsep Utama
- A⁻¹
- A · A⁻¹ = I

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
A = np.array([[2,1],
              [1,1]])

invA = np.linalg.inv(A)
I = A @ invA

print("Inverse:\n", invA)
print("Identity:\n", I)
```

```python
import matplotlib.pyplot as plt

plt.imshow(I)
plt.colorbar()
plt.title("Identity Matrix Visualization")
plt.show()
```

```python
v = np.array([1,2])

v_t = A @ v
v_back = invA @ v_t

plt.quiver(0,0,v[0],v[1], angles='xy', scale_units='xy', scale=1, label='Original')
plt.quiver(0,0,v_t[0],v_t[1], angles='xy', scale_units='xy', scale=1, label='Transformed')
plt.quiver(0,0,v_back[0],v_back[1], angles='xy', scale_units='xy', scale=1, label='Recovered')

plt.legend(); plt.grid(); plt.title("Inverse Transformation"); plt.show()
```

Inverse penting untuk membalik transformasi dan solving system.

# Chapter 3 - Vector Operations & Visualization

## Summary
Chapter ini membahas operasi lanjutan pada vektor serta visualisasi untuk memahami hubungan antar vektor.

## Tujuan
- Memahami panjang (magnitude) vektor
- Memahami arah vektor
- Melakukan visualisasi vektor

## Konsep Dasar

Beberapa konsep penting:
- Magnitude: panjang vektor
- Direction: arah vektor
- Dot product: hubungan antar vektor

```python
import numpy as np

v = np.array([3, 4])

magnitude = np.linalg.norm(v)
print("Magnitude:", magnitude)
```

```python
import matplotlib.pyplot as plt

v = np.array([3, 4])

plt.quiver(0, 0, v[0], v[1], angles='xy', scale_units='xy', scale=1)

plt.text(v[0], v[1], f'({v[0]}, {v[1]})')

plt.xlim(0,5)
plt.ylim(0,5)

plt.title("Vector Visualization with Magnitude")
plt.grid()

plt.show()
```

## Dot Product

Dot product digunakan untuk mengukur hubungan antara dua vektor.

```python
v1 = np.array([1,2])
v2 = np.array([2,1])

dot = np.dot(v1, v2)
print("Dot Product:", dot)
```

```python
v1 = np.array([1,2])
v2 = np.array([2,1])

plt.quiver(0,0,v1[0],v1[1], angles='xy', scale_units='xy', scale=1, label='v1')
plt.quiver(0,0,v2[0],v2[1], angles='xy', scale_units='xy', scale=1, label='v2')

plt.xlim(0,3)
plt.ylim(0,3)

plt.title("Comparison of Two Vectors")
plt.grid()

plt.legend()
plt.show()
```

## Kesimpulan

Operasi vektor seperti magnitude dan dot product sangat penting dalam memahami hubungan data.
Visualisasi membantu memahami arah dan hubungan antar vektor secara intuitif.

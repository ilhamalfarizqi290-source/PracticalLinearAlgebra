# Chapter 10 - Row Reduction (RREF)

## Summary
Row reduction digunakan untuk menyederhanakan sistem persamaan.

## Konsep Utama
- Gaussian elimination
- RREF

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sympy import Matrix

A = Matrix([[1,2,3],
            [2,4,6],
            [1,1,1]])

rref = A.rref()
print(rref)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sympy import Matrix

A = Matrix([[1,2,3],
            [2,4,6],
            [1,1,1]])

rref = A.rref()

# BEFORE
A_np = np.array(A).astype(float)

plt.imshow(A_np)
plt.colorbar()
plt.title("Before RREF")
plt.show()

# AFTER
R = np.array(rref[0]).astype(float)

plt.imshow(R)
plt.colorbar()
plt.title("After RREF")
plt.show()
```

Row reduction membantu menyelesaikan sistem linear secara sistematis.
